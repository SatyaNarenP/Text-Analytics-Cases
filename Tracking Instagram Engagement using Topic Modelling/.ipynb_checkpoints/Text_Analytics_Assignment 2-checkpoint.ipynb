```python
import pickle
import pandas as pd
import numpy as np

post_info = pickle.load( open( "save.p", "rb" ) )
```

# Task A

## Engagement Score

```python
post_info['likes_normalized'] = post_info['likes']/post_info['likes'].max()
```

```python
post_info['comments_normalized'] = post_info['comments']/post_info['comments'].max()
```

```python
post_info['engagement_score'] = .4*post_info['likes_normalized']+.6*post_info['comments_normalized']
```

```python
engagement_median = post_info['engagement_score'].median()

def binary_engage(score):
    if score > engagement_median:
        return 1
    else:
        return 0
```

```python
post_info['engagement_binary'] = post_info['engagement_score'].map(binary_engage)
```

```python
post_info.engagement_binary.value_counts()
```

# Task B

Conclusion: Our best accuracy (78%) was given by regression on just the TF-IDF scores for the captions. Using both the labels and captions (with TF-IDF) actually reduced our accuracy. We can conclude that this additional information introduced either collinearity or noise that reduced accuracy. 

## Logistic Regression (Labels only)

```python
regression = post_info[['labels', 'engagement_binary']]
```

```python
labels = list(set(post_info['labels'].sum()))
```

```python
for label in labels:
    regression[label] = regression['labels'].apply(lambda x: 1 if label in x else 0 )
```

```python
regression
```

```python
X = regression.drop(['labels', 'engagement_binary'], axis = 1)
y = regression[['engagement_binary']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
log_reg.score(x_test, y_test)
```

```python
from sklearn import metrics

predictions = log_reg.predict(x_test)

metrics.confusion_matrix(y_test, predictions)
```

```python
predictions
```

```python
confusion = pd.DataFrame(zip(y_test.engagement_binary, predictions), columns=['actual', 'predicted'])
pd.crosstab(confusion.actual, confusion.predicted)
```

## TF-IDF on Captions

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')

class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc)]
```

```python
cv=CountVectorizer(tokenizer=LemmaTokenizer(), ngram_range=(1,2), stop_words='english', max_features=1250)

word_count_vector=cv.fit_transform(post_info.caption.values)

tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count_vector)

# tf-idf scores
tf_idf_vector=tfidf_transformer.transform(word_count_vector)
```

```python
tf_idf_vector
```

```python
import scipy
from scipy import sparse
```

```python
X = tf_idf_vector

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)

log_reg.score(x_test, y_test)
```

```python
predictions = log_reg.predict(x_test)

confusion = pd.DataFrame(zip(y_test.engagement_binary, predictions), columns=['actual', 'predicted'])
pd.crosstab(confusion.actual, confusion.predicted)
```

# Task C

## Topic Modeling

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.test.utils import common_corpus, common_dictionary
```

```python
dictionary = gensim.corpora.Dictionary(post_info.labels)
dictionary.filter_extremes(no_below=1, no_above=0.5) #, keep_n=100000)
```

```python
bow_corpus = [dictionary.doc2bow(label) for label in post_info.labels]
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=4, id2word=dictionary, passes=2, workers=2, chunksize=100, random_state=3)
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
pp.pprint(lda_model.print_topics())
```

## Topic Interpretation

0. Geology
1. Marine Biology
2. Landscape
3. Wildlife

```python
post_topics = post_info.copy(deep=True)

post_topics['topic_0'] = 0.0
post_topics['topic_1'] = 0.0
post_topics['topic_2'] = 0.0
post_topics['topic_3'] = 0.0
```

```python
for index, row in post_topics.iterrows():
    bow = dictionary.doc2bow(row['labels'])
    t = lda_model.get_document_topics(bow)
    post_topics['topic_0'].loc[index] = t[0][1]
    post_topics['topic_1'].loc[index] = t[1][1]
    post_topics['topic_2'].loc[index] = t[2][1]
    post_topics['topic_3'].loc[index] = t[3][1]
```

```python
labels = list(set([i for l in post_info.labels for i in l]))

word_loadings = pd.DataFrame(labels, columns=['word'])
word_loadings['topic_0'] = 0.0
word_loadings['topic_1'] = 0.0
word_loadings['topic_2'] = 0.0
word_loadings['topic_3'] = 0.0
```

```python
for index, row in word_loadings.iterrows():
  try:
    w = lda_model.get_term_topics(row['word'], minimum_probability=0.000001)
    word_loadings['topic_0'].loc[index]  = w[0][1]
    word_loadings['topic_1'].loc[index]  = w[1][1]
    word_loadings['topic_2'].loc[index]  = w[2][1]
    word_loadings['topic_3'].loc[index]  = w[3][1]
  except:
    pass
```

## Quartiles

```python
post_topics['quartile'] = pd.qcut(post_topics['engagement_score'], 4, labels=False)
```

```python
post_topics.groupby('quartile')[['topic_0','topic_1','topic_2','topic_3']].mean()
```

## Task D

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(post_topics.topic_0, post_topics.engagement_score, alpha = 0.2, label = 'Geology')
plt.scatter(post_topics.topic_1, post_topics.engagement_score, alpha = 0.2, label = 'Marine')
plt.scatter(post_topics.topic_2, post_topics.engagement_score, alpha = 0.2, label = 'Landscape')
plt.scatter(post_topics.topic_3, post_topics.engagement_score, alpha = 0.2, label = 'Wildlife')
plt.legend(loc='best')
plt.xlabel('Topic Weight')
plt.ylabel('Engagement')
```

Based on the results, there is no clear trend showing that just one topic would lead to an increase in engagement. Topic 3 (Wildlife) exhibits the highest engagement followed by Topic 0 (Geology).

While there is no clear trend for a singular topic, there is a clear relationship between the intensity of the topic and engagement. As the LDA topic weight approaches 1 (or equivalently zero), the engagement significantly increases.
